# While Loop


```python
i = 1
while i <= 10:
    print(i)
    i += 1

print( "value of i =", i)
    
```

```python
while True:
    print("While loop")
```

```python
i = 1
sum = 0 
while i <= 10:
    sum = sum + i
    i += 1
    
print(sum)
```

# While Else

```python
i = 11
while i <= 10:
    print(i)
    i += 1
else:
    print("End of while")
```
